### What's a Fourier Transform?

Fourier Transform (FT) is a model to trasform signals between two different domains, from frequency to time domain or vice versa, these is applied in different fields of studies, for example in Physics and engineering. 

These trasformation in domains have different purpusses such as delating a high tun in a  signal (Like music) or making more accurated radar data. In Geo science we used signals to decode information underground, one example of these are the Ground Penetration Radar (GPR) that can find objects with accurecy from centimeter up to 30 meters under ground. In must of the cases we have a signal that is a combination of differents frequencys, making a combination of what is usefull and what is not. We use FT to remove the useless data.


To undestand what is the Fourier Transform we need to start by undestanding what is a signal.

Below it can be see a graph comparing the behavior of a 150 Hz Vs a 75 Hz  (Natural signals) 

```python
"Graph of Simple Harmonic Motion in Time"
import matplotlib.pyplot as plt 
import math 
import numpy as np

f=150
f2=f/2
T=2*math.pi/f
A=1
time=0.2

t= np.linspace(0.0, time,num=160)

x1 = A* np.cos(f*t)
x2 = A* np.cos(f2*t)
y1, y2 = [.1,.1], [ 0,A* np.cos(f*.1)]

p=A* np.cos(f*.1)

plt.subplot(2, 1, 1)
plt.plot(t, x1, '-b.', y1, y2,'k-o')
plt.title('Amplitud VS Time')
plt.ylabel("%.0f Hz" %f)
plt.grid(axis="y")
plt.annotate('value= %.2f'%p,xy=(.2,p),xycoords='data',
            xytext=(50, 1), textcoords='offset points',
            arrowprops=dict(facecolor='black', shrink=.1),
            horizontalalignment='center', verticalalignment='center')
    
p2=A* np.cos(f2*.1)
y1, y2 = [.1,.1], [ 0,A* np.cos(f2*.1)]
plt.subplot(2, 1, 2)
plt.plot(t, x2,'-r.',y1, y2,'k-o' )
plt.ylabel("%.0f Hz" %f2)
plt.grid(axis="y")
plt.annotate('value= %.2f'%p2,xy=(.2,p2),xycoords='data',
            xytext=(50, 1), textcoords='offset points',
            arrowprops=dict(facecolor='black', shrink=.1),
            horizontalalignment='center', verticalalignment='center')
#plt.xlabel("time (s)")

plt.show()
```

# How to find a natural signals of a given complex signal?

Know that we undestand how a complex signal is compused, we can undestand the used of a fourier transform, we appling it we can start with a given signal and find the natural signals that composed it. Allowing us to removed error data, or useless frequencys. 




```python
# Making a complex signal
samplingFrequency   =250;

# At what intervals time points are sampled
samplingInterval       = 1 / samplingFrequency;

# Begin time period of the signals
beginTime           = 0;

# End time period of the signals
endTime             = 2;  

# Frequency of the signals
signal1Frequency     = 5;
signal2Frequency     = 14;
signal3Frequency     = 7;
signal4Frequency     = 25;

# Time points
time        = np.arange(beginTime, endTime, samplingInterval);

# Create two sine waves
amplitude1 = np.cos(2*np.pi*signal1Frequency*time)
amplitude2 = np.cos(2*np.pi*signal2Frequency*time)
amplitude3 = np.cos(2*np.pi*signal3Frequency*time)
amplitude4 = np.cos(2*np.pi*signal4Frequency*time)

# Create subplot
plt.figure(figsize=(21, 7))

# Add the sine waves
amplitude = amplitude1 + amplitude2 +amplitude3 +amplitude4

# Time domain representation of the resultant sine wave
plt.title('Complex signals',fontsize=25)
plt.plot(time, amplitude)
plt.xlabel('Time')
plt.ylabel('Amplitude') 

```

```python
# Applying Fourier Trasform 

# Frequency domain representation
fourierTransform = np.fft.fft(amplitude)/len(amplitude)           # Normalize amplitude
fourierTransform = fourierTransform[range(int(len(amplitude)/2))] # Exclude sampling frequency

tpCount     = len(amplitude)
values      = np.arange(int(tpCount/2))
timePeriod  = tpCount/samplingFrequency
frequencies = values/timePeriod

#Find up to where plot for correct visualitation. 
ten_porcentaje=max(fourierTransform)/10
value=min(fourierTransform)
index=0
for i in range(0,len(fourierTransform)):
    if fourierTransform[i] > ten_porcentaje:
        index=i

# Frequency domain representation
plt.figure(figsize=(21, 7))
plt.title('Fourier transform depicting the frequency components',fontsize=25)
plt.plot(frequencies, abs(fourierTransform))
plt.xlabel('Frequency')
plt.ylabel('Amplitude')
plt.xlim((min(frequencies), (frequencies[index+int(len(frequencies)/10)])))
```

The picks that aprears in the graph Frequency VS Amplitud is the result of applying Fourier Trasform. From the graph we interpret that the complex siganl is form from 4 diferents natural singals.
